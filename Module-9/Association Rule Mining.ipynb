```python
import pandas as pd
import numpy as np
```

```python
file_data = open("store_data.csv",'rt')
store_data = file_data.read()
file_data.close()
```

```python
len(store_data)
```

```python
store_data = store_data.split("\n")
store_data[:5]
```

```python
store_data_final = [element.split(",") for element in store_data]
store_data_final[:5]
```

```python
!pip install mlxtend
```

```python
# Pre-requisite - get the data as a list inside a list
from mlxtend.preprocessing import TransactionEncoder
```

```python
encoder = TransactionEncoder()
encoder.fit(store_data_final)
```

```python
encoder.columns_ # get the unique elements from transaction data
```

```python
transformed_data = encoder.transform(store_data_final)
transformed_data[:5]
```

```python
len(encoder.columns_)
```

```python
transformed_data = transformed_data.astype('int8')
```

```python
import pandas as pd
transformed_df = pd.DataFrame(transformed_data, columns=encoder.columns_)
transformed_df.head()
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemsets = apriori(transformed_df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.25)
rules.sort_values('lift',ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=3)
rules.sort_values('lift',ascending=False)  
```
