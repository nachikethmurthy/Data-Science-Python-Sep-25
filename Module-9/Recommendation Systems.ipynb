```python
# User based collaborative Filtering
# rows - userid
# columns - movieID
# values - rating
user_df = ratings_df.pivot(index='userId', columns='movieId',values='rating')
user_df.head()
```

```python
user_df.fillna(value=0, inplace=True)
user_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_sim = cosine_similarity(user_df)
```

```python
user_sim.shape
```

```python
user_sim[:2]
```

```python
user_sim_df = pd.DataFrame(user_sim, columns=user_df.index, index = user_df.index)
user_sim_df.head()
```

```python
np.fill_diagonal(user_sim,0)
```

```python
movies_df = ratings_df.pivot(columns='userId', index='movieId',values='rating')
movies_df.head()
```

```python
movies_df = ratings_df.pivot(columns='userId', index='movieId',values='rating')
movies_df.fillna(value=0, inplace=True)
movies_sim = cosine_similarity(movies_df)
np.fill_diagonal(movies_sim,0)
movies_sim_df = pd.DataFrame(movies_sim, columns=movies_df.index, index = movies_df.index)
movies_sim_df.head()
```
